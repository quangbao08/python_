```python
import tkinter as tk
from tkinter import ttk
from tkinter import ttk, Label, Tk, Entry, Button,messagebox

# Hàm thêm nhân viên vào Treeview
def them_nhan_vien():
    ma_nv = entry_ma_nv.get()
    ten_nv = entry_ten_nv.get()
    que = entry_que.get()
    so_dien_thoai = entry_so_dien_thoai.get()

    # Thêm thông tin nhân viên vào Treeview
    treeview_data.insert("", "end", values=(ma_nv, ten_nv, que, so_dien_thoai))

    # Xóa nội dung các Entry widgets
    entry_ma_nv.delete(0, tk.END)
    entry_ten_nv.delete(0, tk.END)
    entry_que.delete(0, tk.END)
    entry_so_dien_thoai.delete(0, tk.END)

# Hàm xóa nhân viên từ Treeview
def xoa_nhan_vien():
    data = treeview_data.selection()
    if data:
        treeview_data.delete(data)

# Hàm sửa nhân viên trong Treeview
def sua_nhan_vien():
    data = treeview_data.selection()
    if data:
        values = treeview_data.item(data, "values")
        if values:
            entry_ma_nv.delete(0, tk.END)
            entry_ten_nv.delete(0, tk.END)
            entry_que.delete(0, tk.END)
            entry_so_dien_thoai.delete(0, tk.END)
            entry_ma_nv.insert(0, values[0])
            entry_ten_nv.insert(0, values[1])
            entry_que.insert(0, values[2])
            entry_so_dien_thoai.insert(0, values[3])
            
            
def xem_nhan_vien():
    data = treeview_data.selection()
    if data:
        values = treeview_data.item(data, "values")
        if values:
            thong_tin = f"Mã NV: {values[0]}\nTên NV: {values[1]}\nĐịa Chỉ: {values[2]}\nSố Điện Thoại: {values[3]}"
            # Hiển thị thông tin trong MessageBox
            messagebox.showinfo("Thông Tin Nhân Viên", thong_tin)
        else:
            messagebox.showwarning("Cảnh Báo", "Dòng không chứa giá trị.")
    else:
        messagebox.showwarning("Cảnh Báo", "Không có dòng nào được chọn.")

# Tạo cửa sổ
my_form = tk.Tk()
my_form.title('Quản Lý Nhân Sự')

# Tạo Treeview
listCols = ('Mã NV', 'Tên NV', 'QUÊ QUÁN', 'Số Điện Thoại')
treeview_data = ttk.Treeview(master=my_form, columns=listCols, show='headings')

for col in listCols:
    treeview_data.heading(col, text=col)
    treeview_data.column(col, width=150)

treeview_data.grid(row=0, column=0, columnspan=4, padx=10, pady=10)

# Tạo các Label và Entry widgets để nhập thông tin nhân viên
tk.Label(my_form, text='Mã NV:').grid(row=1, column=0)
entry_ma_nv = tk.Entry(my_form, width=20)
entry_ma_nv.grid(row=1, column=1)

tk.Label(my_form, text='Tên NV:').grid(row=1, column=2)
entry_ten_nv = tk.Entry(my_form, width=20)
entry_ten_nv.grid(row=1, column=3)

tk.Label(my_form, text='QUÊ QUÁN:').grid(row=2, column=0)
entry_que = tk.Entry(my_form, width=20)
entry_que.grid(row=2, column=1)

tk.Label(my_form, text='Số Điện Thoại:').grid(row=2, column=2)
entry_so_dien_thoai = tk.Entry(my_form, width=20)
entry_so_dien_thoai.grid(row=2, column=3)

# Tạo Combobox chứa các chức năng
chuc_nang_combobox = ttk.Combobox(my_form, values=('Thêm', 'Xem', 'Xóa', 'Sửa'))
chuc_nang_combobox.set('Chọn Chức Năng')
chuc_nang_combobox.grid(row=3, column=0, columnspan=2, pady=10)

# Tạo các nút chức năng
tk.Button(my_form, text='OK', command=lambda: xu_ly_chuc_nang(chuc_nang_combobox.get())).grid(row=3, column=2, columnspan=2, pady=10)

# Hàm xử lý chức năng dựa trên giá trị chọn từ Combobox
def xu_ly_chuc_nang(chuc_nang):
    if chuc_nang == 'Thêm':
        them_nhan_vien()
    elif chuc_nang == 'Xem':
        xem_nhan_vien()
    elif chuc_nang == 'Xóa':
        xoa_nhan_vien()
    elif chuc_nang == 'Sửa':
        sua_nhan_vien()
    

# Chạy ứng dụng
my_form.mainloop()
```
